```python
                            #Predicting Wine type

#Importing red and white wine data sets

import pandas as pd
white=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep=';')
white.head()
```

```python
red=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=';')
red.head()
```

```python
#Scatter plot of alcohol vs volatile acidity based on quality.

np.random.seed(700)
redlabels=np.unique(red['quality'])
whitelabels=np.unique(white['quality'])
redcolors=np.random.rand(6,4)
whitecolors=np.append(redcolors,np.random.rand(1,4),axis=0)


for i in range(len(redcolors)):
    redy=red['alcohol'][red['quality']==redlabels[i]]
    redx=red['volatile acidity'][red['quality']==redlabels[i]]
    plt.scatter(redx,redy,c=redcolors[i])
    
    
plt.xlabel('Volatile Acidity');
plt.ylabel('Alcohol');
plt.title('Volatile Acidity vs Alcohol for Red Wine');
```

```python
np.random.seed(700)
for i in range(len(whitecolors)):
    whitex=white['alcohol'][white['quality']==whitelabels[i]]
    whitey=white['volatile acidity'][white['quality']==whitelabels[i]]
    plt.scatter(whitex,whitey,c=whitecolors[i])

plt.xlabel('Volatile Acidity');
plt.ylabel('Alcohol');
plt.title('Alcohol vs Volatile Acidity for White Wine');
```

```python
#Combining white and red wine datasets

red['type']=1
white['type']=0
wines=red.append(white,ignore_index=True)
wines.head()
```

```python
#Correlation Table

import seaborn as sns
corr=wines.corr()
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values)
plt.show()
```

```python
#Splitting the dataset into train and test dataset

from sklearn.model_selection import train_test_split
X=wines.ix[:,0:11]
y=np.ravel(wines['type'])
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
#Standardizing the dataset since the values lies far apart

from sklearn.preprocessing import StandardScaler

#Defining the Scaler

scaler=StandardScaler().fit(X_train)

#Scale the train dataset

X_train=scaler.transform(X_train)

#Scale the test dataset

X_test=scaler.transform(X_test)
```

```python
#Modelling the data


from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(12,activation='relu',input_shape=(11,)))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.output_shape
```

```python
model.summary()
```

```python
model.get_config()
```

```python
model.get_weights()
```

```python
#Compiling and fitting the model

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=20,batch_size=1,verbose=1)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred[:5]
```

```python
#Evaluating the model

score=model.evaluate(X_test,y_test,verbose=1)
print(score)
```
